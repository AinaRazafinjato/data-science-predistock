```python
import pandas as pd
```

```python
# Data Load
csv_path = '../../../data/raw/mouvements_stock_fictifs.csv'
data = pd.read_csv(csv_path)
data.head()
```

```python
# Data Preprocessing
data.info()
```

```python
# Nettoyage des donn√©es
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
data['quantity'] = pd.to_numeric(data['quantity'], errors='coerce')

data.info()
```

```python
table = pd.pivot_table(data, values='quantity', index='date',
                       columns=['product_name'], aggfunc="sum")
table = table.fillna(0)
table
```

```python
import os

# Cr√©ation du dossier de sortie si besoin
output_dir = os.path.join(os.getcwd(), '../../../data/processed')
os.makedirs(output_dir, exist_ok=True)

# Export the cleaned data
data.to_csv(os.path.join(output_dir, 'mouvements_stock_fictifs_cleaned.csv'), index=False)
table.to_csv(os.path.join(output_dir, 'pivot_table.csv'), index=True)
```
